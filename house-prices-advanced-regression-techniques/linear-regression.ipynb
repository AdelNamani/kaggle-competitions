```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt 
```

```python
df_train = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
df_train = df_train.drop(['Id'],axis=1).fillna(method='ffill').fillna(method='bfill')
df_train
```

```python
X_houses_full = df_train.iloc[:,:-1].values
Y_houses_full = df_train.iloc[:,-1].values
X_houses_full, Y_houses_full
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_houses_full, Y_houses_full, test_size=0.20)
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
X_train_onehot = enc.fit_transform(X_train)
X_test_onehot = enc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train_onehot, Y_train)
Y_predict = reg.predict(X_test_onehot)
print('Score test = ' , reg.score(X_test_onehot, Y_predict))
print('Score train = ' , reg.score(X_train_onehot, Y_train))
```

```python
reg = LinearRegression().fit(X_train_onehot, Y_train)

df_test = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv').fillna(method='ffill').fillna(method='bfill')

df_submission = df_test[['Id']]

df_test = df_test.drop(['Id'],axis=1)

X_full_test = np.array(df_test.values)
X_full_test_ordinal = enc.fit_transform(X_full_test)

Y_full_test = reg.predict(X_full_test_ordinal)

df_submission['SalePrice'] = pd.Series(Y_full_test)
df_submission.to_csv('house_prices_submission.csv',index=False)
```
