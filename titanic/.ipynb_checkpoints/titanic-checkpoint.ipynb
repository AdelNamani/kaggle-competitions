```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('./train.csv')
df.head()
```

```python
df = df.drop(['Ticket','Name','PassengerId','Cabin'],axis=1).dropna()
df
```

```python
X_titanic = df.iloc[:,1:].values
Y_titanic = df.iloc[:,0].values
```

```python
X_titanic = np.array(X_titanic)
Y_titanic = np.array(Y_titanic)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_titanic, Y_titanic, test_size=0.20)
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
X_train_onehot = enc.fit_transform(X_train)
X_test_onehot = enc.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

clf = RandomForestClassifier(n_estimators=100,max_depth=3)
clf.fit(X_train_onehot, Y_train)
Y_predicted = clf.predict(X_test_onehot)
Y_predicted_train = clf.predict(X_train_onehot)

from sklearn.metrics import accuracy_score
print("Test accuracy: " + str(accuracy_score(Y_test, Y_predicted)))
print("Train accuracy: " + str(accuracy_score(Y_train, Y_predicted_train)))

from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_predicted,zero_division=1))
```

```python
acc_train = [] 
acc_test = []
for i in range(1,10):
    clf = RandomForestClassifier(n_estimators=100,max_depth=i)
    clf.fit(X_train_onehot, Y_train)
    Y_predicted = clf.predict(X_test_onehot)
    Y_predicted_train = clf.predict(X_train_onehot)
    acc_test.append(accuracy_score(Y_test, Y_predicted))
    acc_train.append(accuracy_score(Y_train, Y_predicted_train))

from matplotlib.legend_handler import HandlerLine2D
line1, = plt.plot(range(1,10), acc_train, 'b', label='Train Accuracy')
line2, = plt.plot(range(1,10), acc_test, 'r', label='Test Accuracy')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('Accuracy score')
plt.xlabel('max depth')
plt.show()
```
